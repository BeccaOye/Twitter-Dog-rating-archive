### Code

```python
#Creating a function to drop columns 
def drop(df, col_name, axis = 1,  inplace=True):
    result = df.drop(col_name, axis, inplace=True)
    return result
```

### 3. The stages of each dog breed

```python
# Making dog stage count a dataframe
d = We_rate_dogs.query('doggo==True').dog_breed.value_counts().rename_axis('breed').reset_index(name='doggo')
f = We_rate_dogs.query('floofer==True').dog_breed.value_counts().rename_axis('breed').reset_index(name='floofer')
pu = We_rate_dogs.query('pupper==True').dog_breed.value_counts().rename_axis('breed').reset_index(name='pupper')
po = We_rate_dogs.query('puppo==True').dog_breed.value_counts().rename_axis('breed').reset_index(name='puppo')

# merge all dog stages
merge1 =  pd.merge(d,f, how='left', on='breed')
merge2 =  pd.merge(pu,po, how='left', on='breed')
dog_stage = pd.merge(merge1,merge2,how='right',on='breed').fillna(0)
dog_stage
```

```python
dog_stage.head(60)
```

Here, we see most of these dogs do not really have fur, however there are more doggo and pupper than other stages.

### 4. Dog with highest rating (numerator)

```python
# denominator ratings equal to 10
dog_rate = We_rate_dogs.query('rating_denominator == 10').groupby('dog_breed').rating_numerator.mean().sort_values()
dog_rate = dog_rate.rename_axis('breed').reset_index(name='average_rate')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
x1 = dog_rate.iloc[:30]
x2 = dog_rate.iloc[30:60]
x3 = dog_rate.iloc[60:90]
x4 = dog_rate.iloc[90:]
plt.figure(figsize = [15,5])

#Plot 1
plt.subplot(1,2,1)
plt.bar(data=x1, x='breed', height='average_rate')
plt.xticks(rotation=90)
plt.xlabel('breeds')
plt.ylabel('rate')
plt.ylim([0,30])

#Plot 2
plt.subplot(1,2,2)
plt.bar(data=x2, x='breed', height='average_rate')
plt.xticks(rotation=90)
plt.xlabel('breeds')
plt.ylabel('rate')
plt.ylim([0,30])


plt.suptitle('Dog Rating')
plt.show()
```

```python
plt.figure(figsize = [15,5])

#Plot 1
plt.subplot(1,2,1)
plt.bar(data=x3, x='breed', height='average_rate')
plt.xticks(rotation=90)
plt.xlabel('breeds')
plt.ylabel('rate')
plt.ylim([0,30])

#Plot 2
plt.subplot(1,2,2)
plt.bar(data=x4, x='breed', height='average_rate')
plt.xticks(rotation=90)
plt.xlabel('breeds')
plt.ylabel('rate')
plt.ylim([0,30])


plt.suptitle('Dog Rating')
plt.show()
```

### What dog breed is the favorite

```python
x=['doggo','floofer','pupper','puppo']
a = (We_rate_dogs.query('doggo==True').favorite_count.mean())
b = (We_rate_dogs.query('floofer==True').favorite_count.mean())
c = (We_rate_dogs.query('pupper==True').favorite_count.mean())
d = (We_rate_dogs.query('puppo==True').favorite_count.mean())

y=np.array([a,b,c,d])
plt.pie(y, labels=x, startangle=90)
plt.title('Count of favorite based on dog stage')
plt.show()
```

This shows puppo as the stage with the highest favorite count
